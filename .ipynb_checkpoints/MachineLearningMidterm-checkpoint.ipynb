Part 1) a. Linear Discriminant Analysis is a method of classification that draws decision boundaries based on the data. If the data falls on one side or section of the data, it will be classified as x or if it falls on the other side, it will be classified as y. The primary goal of LDA is to reduce dimensionality and classify data based on those dimensions. The LDA achieves dimensionality reduction by only selecting a few features from the data to model which it decides by taking the features that provide the most seperability. This is calculated through a score of the differences between the means of the classes and the range of the classes. This is similar but different to PCA. Both PCA and linear discriminant analysis reduce dimensions but they differ in how they choose their features. Linear discrimant analysis will choose the features that provide the most seperability while PCA will choose those with the most variance.

b. The between class scatter matrix and the in class scatter matrix serve an important role. The in class scatter matrix maximizes the difference between the two means of the classes. The between class scatter matrix is the minimizes the variance in the classes themselves. It is important to maximize these for a good LDA. LDA seeks to seperate out the classes while lowering the dimensions and keeping the classification information. In the ratio nof between/within, the more the classes' means are seperate, the igher the score and easier it is to distinguish. For within, the less class variability, the higher the score and better LDA.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA4
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split # Import train_test_split function

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
```

```python
#loaded in the data
data = load_wine()
answers = data.target
pd.set_option("display.max_rows", None, "display.max_columns", None)
# Converting data to a dataframe to view properly
data = pd.DataFrame(data=data['data'],columns=data['feature_names'])
#Fit to an LDA
lda = LDA4(n_components=2)
x = lda.fit_transform(data, answers)

#Plotting the data
plt.figure(figsize=(10, 7))
#Got this code online and not too sure how this works. It plots each point of the LDA accoridng to its class 
for label in range(3):  # Three classes in the Wine dataset
    plt.scatter(x[answers == label, 0], x[answers == label, 1], label=f'Class {label}')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.title('LDA of Wine Dataset')
plt.show()
```

```python
from ISLP import confusion_table
#Split the data into test and training data
X_train, X_test, y_train, y_test = train_test_split(x, answers, test_size=0.3, random_state=0)
# Create a logisitc regression function using the test data
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
# 
y_pred = logReg.predict(X_test)
# Print out a confusion matrix
confusion_table(y_pred, y_test)
```

Part 1) a. Information gain is one of the methods that decisions trees use to decide what feature is best to split on. In a very basic way, it is how much information we can gain from decision. Information gain ranges from 0 to 1 with 0 being the worst or no information gain and 1 being the most information gain possible. This is calculated by taking the entropy of the parent node and subtracting the entropy of the child node. Entropy is a measurement of surprise of drawing a certain outcome from a group of data. I will explain more of this later. Information gain is calculated by taking the node before a split and calculating the entropy of the data. It will then take the data after the split and calculate the entropy of the result. Then it will subtract the entropy of the parent to the entropy of the child. This will be the score of the split. The tree will do this for every feature and select the feature with the best information gain and use this as the split. The tree will continue doing with for however many layers deep it wants to go. Overall, it is a good way to tell how good a split is and how much of an improvement it is over no split.

b. Gini impurity like entropy is a measurement in which splitting decisions are made on. It ranges from 0.5 to 0 with 0.5 being the highest and worst and 0 being the lowest and best. It is defined as 1 - sum of every class ( probabliltiy of each class squared). It sort of measures how "impure" the results are with a more mixed results being more impure and a one sided result being good. A decision tree will use these scores to evaluate the the best feature to split on. It will go through each feature and split on each one. It will then find gini impurity of the split and choose the split with the best gini impurity score. Entropy is similar to gini impurity as it is a score of randomness that decesions trees use. It range from 1 to 0 and with 1 being the worst and 0 being the best. It is calculated through sum((1/log(px)) x p(x)). Trees use this score to decide on the best split possible that results in the most information gain. One difference between the entropy and gini impurity is that gini impurity is simpler because it doesn't use a log which will have faster results. But entropy seems to be more accurate.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.tree import (DecisionTreeClassifier as DTC,
                          DecisionTreeRegressor as DTR,
                          plot_tree,
                          export_text)
from sklearn.metrics import (accuracy_score,
                             log_loss)
from sklearn.ensemble import \
     (RandomForestRegressor as RF,
      GradientBoostingRegressor as GBR)
from ISLP.bart import BART
from ISLP.models import ModelSpec as MS
from ucimlrepo import fetch_ucirepo 
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.preprocessing import LabelEncoder
from matplotlib.pyplot import subplots

# Load the dataset and clean it
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"
#column names gotten from online
column_names = ["poisonous","cap-shape","cap-surface","cap-color","bruises","odor","gill-attachment","gill-spacing	","gill-size","gill-color","stalk-shape","stalk-root","stalk-surface-above-ring","stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring","veil-type","veil-color","ring-number","ring-type	","spore-print-color","population","habitat"]
feature_names = ["cap-shape","cap-surface","cap-color","bruises","odor","gill-attachment","gill-spacing	","gill-size","gill-color","stalk-shape","stalk-root","stalk-surface-above-ring","stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring","veil-type","veil-color","ring-number","ring-type	","spore-print-color","population","habitat"]

#Read the data
data = pd.read_csv(url, header=None, names=column_names)
#Get rid of data that had ? in stalk root. I found this out online that this was the column that had missing data
data = data[data["stalk-root"] != "?"]
#make a target to for the test
target = np.where(data.poisonous == 'p',
                "poison",
                "edible")
data = data.drop(columns = ["poisonous"])
#Change to dummy
for column in data.columns:
    data[column] = pd.factorize(data[column])[0]
# data = pd.get_dummies(data)
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion='entropy',
          max_depth=3,
          random_state=0)

# Train Decision Tree Classifer
clf = clf.fit(data,target)
# The plot
ax = subplots(figsize=(12,12))[1]
print(len(feature_names))
plot_tree(clf,feature_names=feature_names,
          ax=ax);
```

```python
# Make predictions 
y_pred = clf.predict(data)
# Print out a confusion matrix
confusion_table(y_pred, target)
```

Bagging and boosting are both ensemble methods to help build better decisions trees. Ensemeble methods means that there is a group of smaller weaker trees that help drive the decesion process. Bagging (Boostrap Aggregation) is a method in which trees are made from independent subsets of data and run in parellel. To start a sample of data is bootstrapped into multiple smaller subsets. Then each subset is fit with a weak learner or smaller tree. Then when deciding, the ensemble of trees vote together. Boosting is where learners are sequentially one after the other to minimize the bias. When the first tree is fit, it weights the incorrect classification more heavily and fits a new tree. This weighting and fitting continue on until how many trees are specified. Bagging and boosting differ in how they handle bias and variance as bagging will decrease variance through its multiple independent trees while boosting will try and decrease bias by giving more weight to wrong answers to to fit the model as perfectly as possible.

A weak learner is a smaller tree that isn't very good on its own. It is normally limited by the number of layers it can have. The idea of having a bunch of weak learners is that none of them are better than a larger trees by themselves, but with a group of them, they can be more accurate and better than a single tree. The boosting algorithm is an ensemble algorithm which means there are a lot of weak learners that are made. Each iteration of the weighting mistakes and fitting results in a new weak learner. The advantages of weak learners is that a mistake from one of the trees will likely get covered up by the rest of the weak learners

a. Bootstrapping is a method that can allow multiple subsets of data to be created through one set of data. This is done through random sampling through replacement where once a data point has been sampled, it is put back into the pool and can be sampled again. A new subset of data is created through this way. This is used in random forest to create a unqiue subsets of data so each tree can possibly have different splits. Bootstrapping is important so we can unqiue weak learners depending on the data it was given. This ensures we have a variety and reduce variance. 

b. M is seems is the square root of the number of features (p). So the more features the more features there are the more but they are not linearly related.

c. Out of bag data is the data that didn't get used to make the random forest when the data was bootstrapped. To calculate the out of bag error, each out of bag data is fed through the random forest. Each tree in the forest classifies it and votes for what it believes it should be classified. The whole random forest as a whole then decides what the data point should be classified as by voting from the trees. If it is classifed right by the forest, then that is counted as a right classification. If it is wrong, it is counted as an error. For the total error to be calculated, it is the number of errors divided by the total number of out of bag data points that were used. This is similar to cross validation as it takes an instance of the data that is part of the overall data but not used in the training and uses it to test the model.
